```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

num_samples = 200
mean = [-5, 5]
cov = [[1,0], [0,1]]

X = np.zeros((num_samples, 2))

X = np.random.multivariate_normal(mean, cov, num_samples)

print(X.shape)

figure, axes = plt.subplots(1)

plt.scatter(X[:,0], X[:,1], marker='o', color ='red')
```

```python
num_samples = 400

mean_1 = [2, 2]
cov_1 = [[3,0], [0,3]]
X_1 = np.zeros((int(num_samples/2), 2))
X_1 = np.random.multivariate_normal(mean_1, cov_1, int(num_samples/2))
print(X_1.shape)

mean_2 = [-1, -1]
cov_2 = [[2,0], [0,2]]
X_2 = np.zeros((int(num_samples/2), 2))
X_2 = np.random.multivariate_normal(mean_2, cov_2, int(num_samples/2))
print(X_2.shape)

figure, axes = plt.subplots(1)

plt.scatter(X_1[:,0], X_1[:,1], marker='o', color ='red')
plt.scatter(X_2[:,0], X_2[:,1], marker='x', color ='blue')
```
