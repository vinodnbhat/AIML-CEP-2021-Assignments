```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
animal_weight_data = pd.read_excel('Q1_train.xlsx', index_col = False)
animal_weight_data.head()
```

```python
animal_weight_data
```

```python
n = len(animal_weight_data.index)
print('number of datapoints in dataset:', n)
```

```python
cols = [0, 1, 2, 3]
X_data = animal_weight_data[animal_weight_data.columns[cols]]
X_data
```

```python
X_array = X_data.to_numpy()
X_array
```

```python
X = np.hstack((X_array, np.ones((X_array.shape[0], 1), dtype = X_array.dtype)))
X
```

```python
response_cols = [4]
y = animal_weight_data[animal_weight_data.columns[response_cols]].to_numpy()
y
```

```python
XTX = np.matmul(np.transpose(X), X)
print('XTX shape:', XTX.shape)

XTy = np.matmul(X.T, y)
print('XTy shape:', XTy.shape)

rank = np.linalg.matrix_rank(XTX)
print('Rank:', rank)

beta = np.matmul(np.linalg.inv(XTX), XTy)
print('beta:', beta)
```

```python
e_i_residuals_list = []

for i in range(n):
  x_i = (X[i, :])
  y_i = (y[i])
  y_pred_i = np.dot(x_i, beta)
  e_i = y_i - y_pred_i

  e_i_residuals_list.append(e_i)

plt.scatter(y, e_i_residuals_list, color = 'r')
plt.title('Residual plot')
plt.xlabel('Weight_g')
plt.ylabel('Residuals')
plt.grid()
plt.show()
```

```python
y_bar = 0

for i in range(n):
  y_bar += float(y[i])

y_bar /= n
print('y_bar:', y_bar)

sigma_yy = 0

for i in range(n):
  y_i = float(animal_weight_data.iat[i,4])
  sigma_yy += (y_i - y_bar) ** 2

print('sigma_yy:', sigma_yy)

sum_sq_residuals = 0

for i in range(n):
  x_i = (X[i, :])
  y_i = (y[i])
  y_pred_i = np.dot(x_i, beta)
  e_i = y_i - y_pred_i
  sum_sq_residuals += e_i ** 2

print('sum_sq_residuals:', sum_sq_residuals)

mse = sum_sq_residuals/n
rmse = np.sqrt(mse)
print('RMSE:', rmse)

R2 = 1 - sum_sq_residuals/sigma_yy

print('R^2:', R2)
```
